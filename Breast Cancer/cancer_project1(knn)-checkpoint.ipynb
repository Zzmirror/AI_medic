```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, KFold
```

```python
data = pd.read_csv('Documents\dataR2.csv')
X=data.to_numpy()
data
```

```python
np.random.seed = 0
X = X[np.random.permutation(X.shape[0])]

y = X[:, -1]
X = X[:, :-1]
X = X - X.mean(axis=0) / X.std(axis=0)
```

```python

split = int(X.shape[0] * 0.8)

X_train = X[:split]
y_train = y[:split]

X_test = X[split:]
y_test = y[split:]
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=3 )
clf.fit(X_train, y_train)

print(clf.score(X_test, y_test))
print(clf.score(X_train, y_train))
```

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=5 )
clf.fit(X_train, y_train)

print(clf.score(X_test, y_test))
print(clf.score(X_train, y_train))
```

```python

neighbors=[1 , 2 , 3 , 5 , 7 , 9 , 10 , 11]

for n in neighbors :

    kfold=KFold(n_splits=5)
    val_score = 0
    train_score =0

    
    for train_data , val_data in kfold.split(X , y) :
        train_x = X[train_data]
        train_y = y[train_data]

        val_x= X[val_data]
        val_y = y[val_data]

        clf=KNeighborsClassifier(n_neighbors=n , weights = 'distance')
        clf.fit(train_x , train_y )

        train_score += clf.score(train_x, train_y)
        val_score += clf.score(val_x,val_y)
    
    print('neighbors = ', n)
    print(train_score / 5)
    print(val_score / 5)
    print('------------------')

    
```
